### 1. 번역(언어)

&nbsp;번역은 인공지능의 성배라고 불린다. 문장을 이해하는 능력의 유무가 매끄러운 번역을 완성하는 열쇠이기 때문이다. AI 기반 번역 서비스는 **인공신경망 번역** 이라는 신기술과 결합해 더욱더 높은 정확도를 제공한다. 이 인공신경망 번역은 과거 통계기반 번역(SMT:Statistical Machine Translation)과 달리 AI가 스스로 빅데이터를 학습하고 번역하는 최신 번역 기술이다.

&nbsp;둘 다 머신러닝을 통해 결과물을 내놓는다는 점에선 동일하지만, 범주의 영역에서 차이가 드러난다. 통계기반 번역은 구문 단위로 쪼개어 번역하며 또한 사용 빈도가 낮아 충분한 데이터베이스가 확보되지 못한 언어에서는 낮은 정확도를 보여준다. 하지만 인공신경망 번역은 **스스로 문장을 통째로 파악하고 어순, 의미, 문맥별 의미 차이 등을 반영해 스스로 수정을 거친 후 결과물을 출력** 한다. 이른바 '아침에 일어나 아침을 준비한다.' 에서 'morning'과 'breakfast'를 구분할 수 있다는 점이다. 이 신경망 기계 번역(NMT:Neural Machine Translation)은 이미 2018년에는 번역 정확도의 지표가 되는 '블루 스코어'가 35까지 올라갔다. 35는 의미가 문제없이 전달되며 문장으로도 자연스럽다고 평가할 수 있는 수준이다.

![smtnmt](../images/week_1/smt_vs_nmt.jpg)

&nbsp;2016년 11월 중국에서 열린 세계인터넷대회 사법 정화분과회의에 참석한 강민구 부산지법 위원장은 각국의 법관들이 모인 자리에서 음성인식 번역 앱을 사용했다. 중국권 사람들과 대화할 때는 네이버의 '파파고'를, 영어로 의사소통할 때에는 '구글 번역'을 활용했다. 강 지방법원장은 “법률용어까지 정확하게 번역되는 모습을 보고 중국인 통역사가 깜짝 놀라는 표정을 지었다”고 전했다.

&nbsp;2018년 MS사는 중국어 뉴스 기사를 영어로 번역하는 모델이 '인간 수준'을 실현했다고 발표했다. 여기서 인간 수준은 번역가를 의미한다. 또, 그해 여름에 열린 'WMT 2018'에는 영어를 체코어로 번역한 결과가 '슈퍼 휴먼(인간 능가)'수준이라는 보고가 나왔다. 2019년 8월에는 영어의 독일어 번역이 '슈퍼 휴먼'이라는 보고도 있었다.

&nbsp;실생활에서 쉽게 접할 수 있는 '파파고'는 위에서 한번 언급했듯이 이미 뛰어난 성능을 보여주고 있다. 2017년 파파고를 총괄하는 김준석 리더와의 인터뷰에서 김준석 리더는 "파파고의 현재 번역 정확도는 100점 만점에 65점으로 번역 전문가가 아닌 일반인(80점)보다는 다소 떨어지는 수준"이라며 "기술 발전 속도를 볼 때 약 7년 뒤면 AI 번역이 일반인 수준을 넘어설 것으로 보인다"라고 답했다. 비록 분량의 제한이 있지만, 예능 라이브나 웹툰을 기반으로 한 AI 교육 데이터로 사용해 일상적이고 트렌디한 표현의 번역에 강하다.

### 2. 음성
&nbsp;인공지능의 또 다른 성배는 바로 음성인식이다. 방대한 데이터 처리를 위한 하드웨어의 부재로 이른 연구에도 불구하고 음성인식 기술은 2000년대 중반까지도 쉽게 상용화되지 못했다. 하지만 방대한 데이터도 처리할 수 있고 음성을 인식할 수 있는 마이크도 탑재된 **스마트폰의 등장** 과 함께 음성인식 기술은 빠르게 대중화되기 시작했다. 글로벌 음성인식 시장의 규모는 2016년 26억 달러에서 2021년 159억 달러까지 성장할 것으로 전망된다.

&nbsp;주변에서 가장 흔히 볼 수 있는 음성인식의 대표적인 사례는 삼성의 '빅스비'와 애플의 '시리'일 것이다. 이미 200여 개국에서 서비스 되고있는 빅스비는 사용자와 상호작용으로 습관과 선호도에 기반한 개인맞춤형 서비스를 제공한다. 문맥 이해, 자연어 이해, 대화 관리 기술 적용으로 사용자가 말하는 의미를 정확히 이해하고 적합한 반응을 출력해낸다.

![건방진빅스비](./images/week_1/Bixby.jpg)

&nbsp;시리는 사용자의 편의성에 중점을 둔다. 전화나 문자를 대신해주고, 선호하는 음악 추천이나 음악 관련 정보 외에도 무엇이든 물어보면 답해주는 만능 비서의 역할을 표방한다. 검색과 전송이 빨라 각종 기기에 최적화되었으며, 종단 간 암호화 기술로 데이터 보안도 강하다.

&nbsp;네이버는 AI 플랫폼 '클로바'에 스마트 스피커와 클로바 앱에 음성인식 및 음성합성 기술을 적용했다. 여타 다른 만능 비서를 지향하듯이 음성 인식 기반의 쇼핑, 오디오 콘텐츠. '배달의 민족'을 기반으로 하는 음성 주문 등이 가능하다. 또한, AI 스피커를 중심으로 네이버의 각종 콘텐츠를 이용할 수 있다. 단어만으로 대화가 가능하며 동명이인도 인식한다. 음성 합성기술 nVoice의 HDTS 기술이 적용돼 다양한 목소리 서비스도 가능하다.

&nbsp;마지막으로 LG전자 음성인식 AI 기술, 씽큐는 가전제품, 휴대전화 등 기기 제어 분야에 특화됐다. 상황에 따라 변경되는 운전모드를 음성으로 알려주고, 필요한 정보를 말해주는 '교감형 인공지능'이 특징이다. 고객이 주로 머무는 공간은 물론 실내외 온도, 습도, 공기질 등 생활환경과 고객 사용패턴까지 학습해 적절한 코스로 작동할 뿐만 아니라 사용자에게 음성으로 알려준다.

### 3. 이미지
&nbsp;인공지능이 이미지의 정보를 식별하는 기술은 컴퓨터 비전 분야에서 가장 오랫동안 연구된 기술 중 하나이다. 이미지 인식 기술의 과제는 크게 네 가지로 나뉠 수 있는데, **'카테고리 분류(Classification), 위치 추정(Localization), 객체 검출(Object Detection), 객체 분할(Instance Segmentation)’** 등이 이에 해당한다. 기술의 성숙도가 매우 높아진 지금, 이미 이미지 식별기술은 스포츠, 보안 감시, 의료, 포토서비스 등에서 이용되고 있다.

![이미지 인식 과정](./images/week_1/img_process.jpg)

&nbsp;바쁜 현대인들이 3시간가량 지속하는 야구 경기에 집중하는 것은 다소 어려울 수 있다. 그래서 일부 플랫폼 업체들은 경기 종료 후에 해당 경기의 주요 장면과 요약을 담은 하이라이트 클립을 만들어 제공한다. 전문가가 직접 하이라이트 구간을 선정, 편집하는 수동적인 방법을 이용하는 것은 시간적, 비용적 소모가 크다. 엔씨에서는 이미지 인식 기술을 활용한 야구 하이라이트 클립 자동 생성기술에 대한 프로젝트를 진행해왔다.

&nbsp;또한 이미지 식별기술은 공항 보안검색대에서도 활용될 수 있다. 아직은 경험이 많은 보안요원이 3\~4초 사이에 경험에 근거하여 위험 물질을 판독하지만, 그 정확도가 100%일 수는 없다. 엑스레이 검색 장비는 금속, 액체류 등 수하물의 종류에 따라 3\~6가지의 색으로 구분해 보여주는 정도로 현재 상용화되어있다. 수하물의 이미지를 판독하여 위험도를 알려주는 장비들이 최근 도입 준비 중이다.

![이미지_보안](./images/week_1/img_security.jpg)

&nbsp;성인남녀 사망률 상위를 차지하는 질병, 폐암은 말기에 발견되는 경우 생존율이 6.1%지만 조기 발견의 경우 생존율이 64%까지 상승한다. 하지만 국내 폐암 조기 진단 비율은 20.7%에 불과한데, 그 이유는 병변의 크기가 작거나 다른 장기에 가려져 판독하기 어려운 X-선 검사에서 폐암을 놓쳐서 그러하다. 이미지 식별 기술을 도입해 CT 검사 결과를 기반으로 학습한 이미지 분석 엔진이 폐 결절을 97% 정확도로 찾아내고, 또 영상 판독을 보조해주는 경우 의사의 판독 정확도는 20%가 향상되는 실험 결과를 내놓았다. 폐암뿐만 아니라 유방암이나 조직 검사, 그리고 다른 질병에도 활용 범위를 넓혀가고 있고 국내 주요 대학병원에서 이미 활용 중이다.

![이미지_의료](./images/week_1/img_medical.jpg)

&nbsp;실생활에서 가장 흔하게 접할 수 있는 이미지 식별기술은 구글 포토 서비스에서 볼 수 있다. 특정 지역에서 찍은 사진을 찾고 싶거나, 특정 인물을 검색하고 싶다면 구글 포토가 스스로 분류해 놓은 사진첩을 통해 관련 사진을 찾을 수 있다. 심지어 강아지가 등장하거나, 우산이 등장하는 사진을 찾고자 할 땐 검색어에 '강아지','우산' 등의 키워드를 입력하면 그에 맞는 결과를 출력해준다. 이는 검색엔진에서도 활용도가 높은 기술이다.

### 4. 자율주행
&nbsp;운전자의 조작 없이 주행이 가능한 자율주행의 시대가 성큼 다가왔다. 자율주행 자동차가 스스로 주행이 가능하게 하려면 **'주변 환경인지', '주행 도로 판단과 선정', '안전한 기능 제어'** 등 세 가지 기능이 반드시 수반되어야 한다. 그리고 **딥러닝 기술은 이 기능 모두에 활용** 될 수 있다. 실제로 여러 곳에서 자율주행 자동차의 인지, 판단, 제어 기능 향상에 딥러닝을 적용하는 연구가 활발하게 진행되고 있다.

&nbsp;테슬라는 2015년에 적응형 크루즈 컨트롤을 Model5에 적용했고, 볼보는 2016년에 자체 조종, 차선 유지 기능을 적용했다. GM과 아우디는 2017년에 자체 조종, 차선 유지 기능과 차량 정체 지원 기능을 적용하였다. 구글은 2016년 기준 가장 많은 운행 기록을 가지고 있으며, 주행 테스트를 하면서 커브길 주행, 충돌 방지, 비상차(경찰차/응급차/소방차 등) 출현 감지 시 경로변경 등의 방대한 빅 데이터를 수집하고 수집된 데이터를 인공지능(머신러닝, 딥러닝)을 활용해 학습하고 지속적으로 업데이트하고 있다.

![자율주행](./images/week_1/autodrive.jpg)

&nbsp;현대자동차와 기아자동차는 운전자의 주행 성향에 맞는 부분 자율주행을 구현하는 기술인 SCC-ML을 세계 최초로 개발했다. 스마트 크루즈 컨트롤 SCC는 앞차와의 거리를 일정하게 유지해 운전자가 설정한 속도로 자율주행을 해주는 기능으로 ADAS(첨단 운전자보조 시스템)의 주요 기술 중 하나다. SCC-ML은 여기에 AI 기술을 더해 운전자의 주행 성향을 차가 스스로 학습해 SCC 작동 시 운전자와 거의 흡사한 패턴으로 자율주행을 해준다.

&nbsp;현재 자율주행 관련 기술은 '국제 자동차 기술자협회(Society of Automotive Engineers International)'가 지난 2016년에 6(0단계~5단계)단계로 분류한 기술을 바탕으로 적용되고 있다. 그 마지막 단계인 5단계는 자동차가 모든 상황을 스스로 판단하고 주행에 반영하는 것을 의미하고, 중간격인 3단계는 주행에 어느 정도의 의지가 반영되는 단계를 의미한다. 현재 상용화되는 단계는 대부분 2단계에 머무르고 있다.

&nbsp; 인공지능이 활용된 기술은 생활 깊숙이 침투해 있다. 단순히 인간의 지적 활동을 보조해 주는 편의성을 넘어 안전과 관련된 영역까지 넘실거린다. 인공지능은 명실상부한 훌륭한 도구이며 인류의 든든한 조언자가 될 것이다. 기술의 혁신과 함께 성숙해야 할 것은 도구를 다루는 인류의 태도이다. 위의 예시에선 언급하지 않았지만 찾아본 사례 중엔 이미지 식별 기술을 이용한 불법 성인 영상을 만드는 기술도 존재했다. 또한 자율주행할 때 발생한 사고의 책임소재에 대한 문제도 있었다. 날이 잘 선 식칼은 훌륭한 주방 도구이지만, 상황과 때에 따라선 위험한 장면을 연출해낸다. 범람하는 인공지능의 파도 속에서 굳건해져야 할 것은 명확해 보인다.
