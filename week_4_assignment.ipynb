### 심층 신경망
: 입력층과 출력층 사이에 여러 개의 은닉층들로 이뤄진 인공신경망이다. 심층 신경망은 일반적으로 인공신경망과 마찬가지로 복잡한 비선형 관계들을 모델링 할 수 있다. 신층신경망의 목적은 분류 및 수치예측을 하기 위함이고 이미지 트레이닝이나 문자인식과 같은 분야에서 매우 유용하게 쓰이고 있다.

### MCP 뉴런
:1943년 맥컬록과 월터 피츠가 맥컬록-피츠(MCP)라 불리는 뇌세포 개념을 발표했다. 뉴런은 뇌의 신경세포와 서로 연결되어 있으며 화학적, 전기적신호를 처리하고 전달하는데 관여한다. 이후 1957년 프랭크 로젠 블랫은 MCP뉴런 모델을 바탕으로 '퍼셉트론'학습개념을 발표한다.

### 퍼셉트론
:1957년 프랭크 로젠 블랫이 발표한 개념으로, 간단히 표현하자면 입력을 받아 계산한 후, 출력을 반환하는 구조로 나타낼 수 있다. 퍼셉트론은 입력과 임계치를 포함한 활성함수, 출력으로 구성되어 있으며 입력값이 활성함수로 입력되게 한다. 이때 입력값의 누적이 임계치보다 크면 1을 반환하고 작으면 0을 반환하는 바이너리 결과를 출력값으로 갖는다. MCP뉴런의 논리게이트를 모델로 구현한것이기 때문에 근본적으로 차이가 없다.

### 역전파
:퍼셉트론은 XOR문제도 풀지 못한다는 사실에 인공신경망 연구가 침체에 빠져있을때, 1987년 역전파가 세상에 공개되었고, 신경망의 부활이 시작됬다. 역전파는 뉴런의 가중치를 효율적으로 조정하기 위하여 거꾸로 무엇인가를 전파하는 방식이다. 역전파는 기존의 출력값을 지우고, 새로운 출력값으로 목표출력값(지도데이터)를 전달한다. 이러한 값이 나오기 위해서는 바로 이전의 값을 무시할 수 없다. 그리고 바로 이전의 값에서 더 큰 값은 자연스럽게 더 큰 영향을 미치게 될것이다. 신경망은 은닉층의 노드값과 고정값을 비교하여 값이 큰쪽의 가중치 혹은 편향을 더 크게 조정하여 출력값을 지도데이터에 맞추도록 조절한다. 결국 신경망이 지도데이터와 출력값사이의 오차로 가중치를 조절하는것이다. 즉, 역전파는 출력값과 지도데이터 사이에 생기는 오차를 이용해 출력층에서 입력층방향쪽으로 가중치를 조절하는 것이다. 이러한 역전파는 경사하강법을 이용한다.

### 강화학습
:강화학습은 지도학습처럼 정답이 있는것도, 비지도학습처럼 데이터만을 기반으로 학습히자 않는다. 강화학습은 에이전트라는 존재가 환경과 상호작용 하며 이 환경에는 보상이라는 기준이 있어서 다양한 시행착오를 겪어가며 보상을 최대화 하는 방향으로 학습을 진행한다. 강화학습을 한문장으로 정리해보면

'보상을 최대화 하는 의사결정 전략'

즉, 순차적인 행동을 알아낙는 방법이다. 여기서 순차적으로 계속 행동을 결정해야하는 문제를 수학적으로 정의한것이 그 유명한 Markov decision process(MDP)이다.



### 과적합

```
import tensorflow as tf
import math

tf.compat.v1.set_random_seed(2020)
x = 1
y = 0
w = tf.random.normal([1],0,1)

def sigmoid(x):
  return 1/(1 + math.exp(-x))

for i in range(1000):
  output = sigmoid(x*w)
  error = y - output
  w = w + x * 0.1 * error # 경사하강법

  if i % 100 == 99:
    print("학습횟수:",i,"Error:",error,"예측 결과:",output)
```
